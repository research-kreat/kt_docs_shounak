```python
def get_patent_info(sum_df, patent_df):
    # Create an empty DataFrame to store the merged information
    merged_df = pd.DataFrame(columns=['patent_id', 'patent_type', 'patent_date', 'patent_title', 'patent_abstract', 'num_claims', 'summary_text'])

    # Iterate through each row in sum_df
    for index, row in sum_df.iterrows():
        # Get the patent ID from sum_df
        patent_id = row['patent_id']
        
        # Search for the patent ID in patent_df
        patent_row = patent_df.loc[patent_df['patent_id'] == patent_id]
        
        # If the patent ID is found in patent_df, extract relevant information
        if not patent_row.empty:
            patent_type = patent_row['patent_type'].iloc[0]
            patent_date = patent_row['patent_date'].iloc[0]
            patent_title = patent_row['patent_title'].iloc[0]
            patent_abstract = patent_row['patent_abstract'].iloc[0]
            num_claims = patent_row['num_claims'].iloc[0]
            
            # Store the extracted information along with the summary_text from sum_df in the merged DataFrame
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'patent_type': patent_type,
                'patent_date': patent_date,
                'patent_title': patent_title,
                'patent_abstract': patent_abstract,
                'num_claims': num_claims,
                'summary_text': row['summary_text']
            }, ignore_index=True)
    
    # Return the merged DataFrame
    return merged_df

# Example usage:
merged_df = get_patent_info(summ_st, patent)
merged_df
```

```python

def merge_assignee_info(sum_df, assignee_details):
    # Create an empty DataFrame to store the merged information
    merged_df = pd.DataFrame(columns=['patent_id', 'assignee_name', 'assignee_org'])

    # Iterate through each row in sum_df
    for index, row in sum_df.iterrows():
        # Get the patent ID from sum_df
        patent_id = row['patent_id']
        
        # Search for the patent ID in assignee_details
        assignee_row = assignee_details.loc[assignee_details['patent_id'] == patent_id]
        
        # If the patent ID is found in assignee_details, extract relevant information
        if not assignee_row.empty:
            assignee_first_name = assignee_row['disambig_assignee_individual_name_first'].iloc[0]
            assignee_last_name = assignee_row['disambig_assignee_individual_name_last'].iloc[0]
            assignee_org = assignee_row['disambig_assignee_organization'].iloc[0]
            
            # Concatenate first and last names to form the assignee name
            assignee_name = f"{assignee_first_name} {assignee_last_name}"
            
            # Store the extracted information in the merged DataFrame
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'assignee_name': assignee_name,
                'assignee_org': assignee_org
            }, ignore_index=True)
    
    # Return the merged DataFrame
    return merged_df

# Example usage:
merged_df = merge_assignee_info(summ_st, assignee)
merged_df
```

```python
def merge_inventor_info(sum_df, inventor):
    # Create an empty DataFrame to store the merged information
    merged_df = pd.DataFrame(columns=['patent_id', 'inventors'])

    # Iterate through each row in sum_df
    for index, row in sum_df.iterrows():
        # Get the patent ID from sum_df
        patent_id = row['patent_id']
        
        # Filter inventor DataFrame to rows matching the current patent ID
        inventor_rows = inventor.loc[inventor['patent_id'] == patent_id]
        
        # If inventor rows are found for the current patent ID, extract inventor names
        if not inventor_rows.empty:
            inventor_names = []
            for inv_index, inv_row in inventor_rows.iterrows():
                inventor_first_name = inv_row['disambig_inventor_name_first']
                inventor_last_name = inv_row['disambig_inventor_name_last']
                inventor_name = f"{inventor_first_name} {inventor_last_name}"
                inventor_names.append(inventor_name)
            
            # Join inventor names into a single string
            inventors_str = ', '.join(inventor_names)
            
            # Store the extracted information in the merged DataFrame
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'inventors': inventors_str
            }, ignore_index=True)
    
    # Return the merged DataFrame
    return merged_df

# Example usage:
merged_df = merge_inventor_info(summ_st, inventor)
merged_df
```

```python
def merge_cpc_info(sum_df, cpc):
    
    # Convert 'patent_id' columns to string type
    sum_df['patent_id'] = sum_df['patent_id'].astype(str)
    cpc['patent_id'] = cpc['patent_id'].astype(str)


    # Create an empty DataFrame to store the merged information
    merged_df = pd.DataFrame(columns=['patent_id', 'cpc_groups'])

    # Iterate through each row in sum_df
    for index, row in sum_df.iterrows():
        # Get the patent ID from sum_df
        patent_id = row['patent_id']
        
        # Filter cpc DataFrame to rows matching the current patent ID
        cpc_rows = cpc.loc[cpc['patent_id'] == patent_id]
        
        # If cpc rows are found for the current patent ID, extract CPC groups
        if not cpc_rows.empty:
            # Extract CPC groups as a list
            cpc_groups = cpc_rows['cpc_group'].tolist()
            
            # Join CPC groups into a single string
            cpc_groups_str = ', '.join(cpc_groups)
            
            # Store the extracted information in the merged DataFrame
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'cpc_groups': cpc_groups_str
            }, ignore_index=True)
        else:
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'cpc_groups': None
            }, ignore_index=True)
    
    # Return the merged DataFrame
    return merged_df

# Example usage:
merged_df = merge_cpc_info(summ_st, cpc)
merged_df
```

```python
def merge_ipc_info(sum_df, ipc):
    # Convert 'patent_id' columns to string type
    sum_df['patent_id'] = sum_df['patent_id'].astype(str)
    ipc['patent_id'] = ipc['patent_id'].astype(str)

    # Create an empty DataFrame to store the merged information
    merged_df = pd.DataFrame(columns=['patent_id', 'ipc_groups'])

    # Iterate through each row in sum_df
    for index, row in sum_df.iterrows():
        # Get the patent ID from sum_df
        patent_id = row['patent_id']
        
        # Filter ipc DataFrame to rows matching the current patent ID
        ipc_rows = ipc.loc[ipc['patent_id'] == patent_id]
        
        # If ipc rows are found for the current patent ID, extract IPC groups
        if not ipc_rows.empty:
            # Concatenate specified columns into a single string for each row
            ipc_groups = [''.join(map(str, ipc_row[['section', 'ipc_class', 'subclass', 'main_group', 'subgroup']])) for _, ipc_row in ipc_rows.iterrows()]
            
            # Store the extracted information in the merged DataFrame
            merged_df = merged_df._append({
                'patent_id': patent_id,
                'ipc_groups': ipc_groups
            }, ignore_index=True)
        else:
            print("No IPC data found for patent ID:", patent_id)
    
    # Return the merged DataFrame
    return merged_df

# Example usage:
merged_df = merge_ipc_info(summ_st, ipc)
merged_df
```

```python

# Function to retrieve patent title based on patent ID
def get_patent_title(df, patent_id):
    # Search for the patent ID in the DataFrame
    patent_row = df.loc[df['patent_id'] == patent_id]
    # Check if the patent ID exists
    if not patent_row.empty:
        # Retrieve the patent title from the located row
        patent_title = patent_row['ipc_class'].iloc[0]
        return patent_title
    else:
        return "Patent ID not found in the DataFrame."

# Example usage:
patent_id = 'PP35566' # Replace with the patent ID you want to search for
patent_title = get_patent_title(ipc, patent_id)
print(patent_title)
```

```python
def merge_all_info(sum_df, patent_df, assignee_details, inventor, cpc, ipc):
    # Convert 'patent_id' columns to string type
    sum_df['patent_id'] = sum_df['patent_id'].astype(str)
    patent_df['patent_id'] = patent_df['patent_id'].astype(str)
    assignee_details['patent_id'] = assignee_details['patent_id'].astype(str)
    inventor['patent_id'] = inventor['patent_id'].astype(str)
    cpc['patent_id'] = cpc['patent_id'].astype(str)
    ipc['patent_id'] = ipc['patent_id'].astype(str)

    # Merge patent information
    merged_df = pd.merge(sum_df, patent_df, on='patent_id', how='left')
    
    # Merge assignee information
    merged_df = pd.merge(merged_df, assignee_details, on='patent_id', how='left')
    merged_df['assignee_name'] = merged_df.apply(lambda row: f"{row['disambig_assignee_individual_name_first']} {row['disambig_assignee_individual_name_last']}", axis=1)
    
    # Merge inventor information
    inventor['inventor_name'] = inventor.apply(lambda row: f"{row['disambig_inventor_name_first']} {row['disambig_inventor_name_last']}", axis=1)
    inventor_grouped = inventor.groupby('patent_id')['inventor_name'].apply(list).reset_index(name='inventors')
    merged_df = pd.merge(merged_df, inventor_grouped, on='patent_id', how='left')
    
    # Merge CPC information
    cpc_grouped = cpc.groupby('patent_id')['cpc_group'].apply(lambda x: ', '.join(x)).reset_index(name='cpc_groups')
    merged_df = pd.merge(merged_df, cpc_grouped, on='patent_id', how='left')
    
    # Merge IPC information
    ipc_grouped = ipc.groupby('patent_id')[['section', 'ipc_class', 'subclass', 'main_group', 'subgroup']].apply(lambda x: x.to_dict(orient='records')).reset_index(name='ipc_groups')
    merged_df = pd.merge(merged_df, ipc_grouped, on='patent_id', how='left')
    
    return merged_df

# Example usage:
merged_df = merge_all_info(summ_st, patent, assignee, inventor, cpc, ipc)
merged_df
```
