```python
# Iterate through the DataFrame and insert each row into the MongoDB collection with a progress bar
for index, row in tqdm(patents.iterrows(), total=patents.shape[0], desc='Inserting rows into MongoDB'):
    record = {
        'title': row['patent_title'],
        'knowledge_type': 'patent',
        'patent_id': row['patent_id'],
        'patent_type': row['patent_type'],
        'patent_date': row['patent_date'],
        'patent_abstract': row['patent_abstract'],
        'num_claims': row['num_claims'],
        'assignee_name': row['assignee_name'],
        'assignee_org': row['assignee_org'],
        'inventors': row['inventors'],
        'ipc': row['ipc'],
        'summary_text': row['summary_text']
    }
    collection.insert_one(record)
```

```python
journals = pd.read_csv('journals.csv')
```

```python
def insert_journals_into_mongodb(df, collection):
    for index, row in tqdm(df.iterrows(), total=df.shape[0], desc='Inserting rows into MongoDB'):
        record = {
            'doi': row['doi'],
            'doi_url': row['doi_url'],
            'url': row['url'],
            'url_for_landing_page': row['url_for_landing_page'],
            'url_for_pdf': row['url_for_pdf'],
            'published_date': row['published_date'],
            'publisher': row['publisher'],
            'title': row['title'],
            'journal_name': row['journal_name'],
            'journal_issns': row['journal_issns'],
            'updated': row['updated'],
            'year': row['year'],
            'pdf_text': row['pdf_text'],
            'summary_text': row['summary'],
            'knowledge_type': 'journal'
        }
        collection.insert_one(record)


# Insert DataFrame rows into MongoDB
insert_journals_into_mongodb(journals, collection)

print("Data insertion completed.")
```
